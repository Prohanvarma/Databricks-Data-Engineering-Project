## Understanding Delta Lake Transaction Log

Understand the cloud storage directory structure behind delta lake tables

0. Create a new schema under the demo catalog for this section of the course (delta_lake)

```sql
CREATE CATALOG IF NOT EXISTS demo
MANAGED LOCATION 'abfss://demo@databcextdl.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
    MANAGED LOCATION 'abfss://demo@databcextdl.dfs.core.windows.net/delta_lake';
```

**1. Create a Delta Lake**

```sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies
(company_name STRING,
 founded_date DATE,
 country STRING);
```

```sql
DESC EXTENDED demo.delta_lake.companies
```

2. Insert some data

```sql
INSERT INTO demo.delta_lake.companies
VALUES ("Apple", "1976-04-01", "USA");
```

```sql
select * from demo.delta_lake.companies
```
