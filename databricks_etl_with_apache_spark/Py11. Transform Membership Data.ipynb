## Transform Membership Data

1. Extract customer_id from the file path 
2. Write transformed data to the Silver Schema

```python
df_memberships = spark.table('gizmobox.bronze.py_memberships')
display(df_memberships)
```

1. Extract customer_id from the file path

```python
from pyspark.sql import functions as F
df_extracted_memberships = (
    df_memberships
    .select(
        F.regexp_extract("path", r".*/([0-9]+)\.png$",1).alias("customer_id"),
        F.col("content").alias("membership_card")
    )
)
display(df_extracted_memberships)
```

2. Write transformed data to the Silver Schema

```python
df_extracted_memberships.writeTo("gizmobox.silver.memberships").createOrReplace()
```
