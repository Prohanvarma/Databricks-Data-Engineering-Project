User Defined Function (UDF)

### Syntax

1. CREATE FUNCTION
2. RETURNS
3. RETURN

Write a function to concatenate firstname and surnameto and output the fullname

```python
%sql
CREATE OR REPLACE FUNCTION gizmobox.default.get_payment_status(payment_status INT)
RETURNS STRING
RETURN CASE payment_status
        WHEN 1 THEN 'Success'
        WHEN 2 THEN 'Pending'
        WHEN 3 THEN 'Cancelled'
        WHEN 4 THEN 'Failed'
      END;
      
      
```
