###  Extract Data From the Address Files

1. Read all addresses files
2. Create addresses table in the bronze layer

1. Read all the address files

```python
df = (
    spark.read.format('csv')
         .option('header', 'true')
         .option('delimiter', '\t')
         .load('/Volumes/gizmobox/landing/operational_data/addresses')
)
display(df)
```

2. Create addresses table in bronze layer

```python
df.writeTo('gizmobox.bronze.py_addresses').createOrReplace()
```
