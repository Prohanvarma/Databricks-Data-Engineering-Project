Set-up the project environment for GizmoBox Data Lakehouse

1. Access the container gizmobox

```sql
%fs ls 'abfss://gizmobox@databcextdl.dfs.core.windows.net/landing/operational_data/'
```

2. Create External Location

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS data_bc_ext_dl_gizmobox
    URL 'abfss://gizmobox@databcextdl.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL data_bc_ext_sc)
    COMMENT 'External Location for demo purposes;'
```

3. Create the catalog - gizmobox

```sql
SHOW CATALOGS
```

```sql
CREATE CATALOG IF NOT EXISTS gizmobox
    MANAGED LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/'
    COMMENT 'This is the catalog for Gizmobox Data Lakehouse';
```

4. Create Schemas

    1. Landing
    2. Bronze
    3. Silver
    4. Gold

```sql
SELECT current_catalog();
```

```sql
USE CATALOG gizmobox;

CREATE SCHEMA IF NOT EXISTS landing
    MANAGED LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/landing';
CREATE SCHEMA IF NOT EXISTS bronze
    MANAGED LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS silver
    MANAGED LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS gold
    MANAGED LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/gold';
```

```sql
SHOW SCHEMAS;
```

5. Create Volume

```sql
USE CATALOG gizmobox;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
    LOCATION 'abfss://gizmobox@databcextdl.dfs.core.windows.net/landing/operational_data/';
```
