## Set-up the project environment for Circuitbox Data Lakehouse

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS data_bc_ext_dl_circuitbox
URL 'abfss://circuitbox@databcextdl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL data_bc_ext_sc)
COMMENT 'External location for the circuitbox data lakehouse';
```

```sql
%fs ls 'abfss://circuitbox@databcextdl.dfs.core.windows.net/'
```

2. Create catalog

```sql
CREATE CATALOG IF NOT EXISTS circuitbox
    MANAGED LOCATION 'abfss://circuitbox@databcextdl.dfs.core.windows.net/'
    COMMENT 'Catalog for the circuitbox data lakehouse';
```

3. Create Schemas

```sql
USE CATALOG circuitbox;

CREATE SCHEMA IF NOT EXISTS landing
    MANAGED LOCATION 'abfss://circuitbox@databcextdl.dfs.core.windows.net/landing';

CREATE SCHEMA IF NOT EXISTS lakehouse
    MANAGED LOCATION 'abfss://circuitbox@databcextdl.dfs.core.windows.net/lakehouse';
```

```sql
SHOW SCHEMAS;
```

```sql
USE CATALOG circuitbox;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
    LOCATION 'abfss://circuitbox@databcextdl.dfs.core.windows.net/landing/operational_data';
```

```sql
%fs ls /Volumes/circuitbox/landing/operational_data
```
